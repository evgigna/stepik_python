```python
# –ó–∞–¥–∞—á–∞_2
# –î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (n‚â§ 9). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç —Ç–∞–±–ª–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ó5, –≥–¥–µ –≤ i-–æ–π —Å—Ç—Ä–æ–∫–µ —É–∫–∞–∑–∞–Ω–æ —á–∏—Å–ª–æ i (—á–∏—Å–ª–∞ –æ—Ç–¥–µ–ª–µ–Ω—ã –æ–¥–Ω–∏–º –ø—Ä–æ–±–µ–ª–æ–º).

# –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

# –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–∞–±–ª–∏—Ü—É —Ä–∞–∑–º–µ—Ä–æ–º n√ó5 –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í –∫–æ–Ω—Ü–µ —Å—Ç—Ä–æ–∫–∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–æ–±–µ–ª.

n = int(input())
for i in range(1, n+1):
    for j in range(1, 6):
        print(i, end=" ")
    print()
print()
```

```python
# –ó–∞–¥–∞—á–∞_3

# –î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n (n‚â§ 9). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç —Ç–∞–±–ª–∏—Ü—É —Å–ª–æ–∂–µ–Ω–∏—è –¥–ª—è –≤—Å–µ—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ n –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–º–µ—Ä–æ–º.

# –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.
 
# –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–∞–±–ª–∏—Ü—É —Å–ª–æ–∂–µ–Ω–∏—è –¥–ª—è –≤—Å–µ—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ n.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í –∫–æ–Ω—Ü–µ —Å—Ç—Ä–æ–∫–∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–æ–±–µ–ª.

n = int(input())
for i in range(1, n+1):
    for j in range(1, 10):
        print(i, "+", j, "=", i+j)
    print()
print()
```

```python
# –ó–∞–¥–∞—á–∞_4

# –î–∞–Ω–æ –Ω–µ—á–µ—Ç–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç —Ä–∞–≤–Ω–æ–±–µ–¥—Ä–µ–Ω–Ω—ã–π –∑–≤–µ–∑–¥–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å –æ—Å–Ω–æ–≤–∞–Ω–∏–µ–º, —Ä–∞–≤–Ω—ã–º n –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–º–µ—Ä–æ–º:

# *
# **
# ***
# ****
# ***
# **
# *

# –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–µ—á–µ—Ç–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

# –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª for.

n = int(input())

for i in range(1, n//2+2):
    for j in range(1, i+1):
        print("*", end="")
    print()

for b in range(n//2, 0, -1):
    print("*"*b)
print()
```

```python
# –ó–∞–¥–∞—á–∞_5

# –î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç —á–∏—Å–ª–µ–Ω–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–º–µ—Ä–æ–º:

# 1
# 22
# 333
# 4444
# 55555
# ...

# –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

# –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª for.

n = int(input())

for i in range(n):
    for j in range(i+1):
        print(i+1, end="")
    print()
```

```python
# –ó–∞–¥–∞—á–∞_6

# –†–µ—à–∏—Ç–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞—Ö  28n+30k+31m=365.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª for. –í –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å –∑–∞–ø–∏—à–∏—Ç–µ —Ä–µ—à–µ–Ω–∏–µ —Å –Ω–∞–∏–º–µ–Ω—å—à–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º n.

total = 0
for n in range(14):
    for k in range(13):
        for m in range(12):
            if 28 * n + 30 * k + 31 * m == 365:
                print("n=", n, "k=", k, "m=", m)
                total += 1
print('total=', total)
```

```python
# –ó–∞–¥–∞—á–∞_7

# –°—Ç–∞—Ä–∏–Ω–Ω–∞—è –∑–∞–¥–∞—á–∞

# –ò–º–µ–µ—Ç—Å—è 100 —Ä—É–±–ª–µ–π. –°–∫–æ–ª—å–∫–æ –±—ã–∫–æ–≤, –∫–æ—Ä–æ–≤ –∏ —Ç–µ–ª—è—Ç –º–æ–∂–Ω–æ –∫—É–ø–∏—Ç—å –Ω–∞ –≤—Å–µ —ç—Ç–∏ –¥–µ–Ω—å–≥–∏, –µ—Å–ª–∏ –ø–ª–∞—Ç–∞:
# - –∑–∞ –±—ã–∫–∞ ‚Äì 10 —Ä—É–±–ª–µ–π,
# - –∑–∞ –∫–æ—Ä–æ–≤—É ‚Äì 5 —Ä—É–±–ª–µ–π,
# - –∑–∞ —Ç–µ–ª–µ–Ω–∫–∞ ‚Äì 0.5 —Ä—É–±–ª—è
# –∏ –Ω–∞–¥–æ –∫—É–ø–∏—Ç—å 100 –≥–æ–ª–æ–≤ —Å–∫–æ—Ç–∞?

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª for.

for x in range(11):
    for y in range(21):
        for z in range(201):
            if 10 * x + 5 * y + 0.5 * z == 100 and x + y + z == 100:
                print("–±—ã–∫–æ–≤", x, "–∫–æ—Ä–æ–≤", y, "—Ç–µ–ª—è—Ç", z)
```

```python
# –ó–∞–¥–∞—á–∞_8

# –ì–∏–ø–æ—Ç–µ–∑–∞ –≠–π–ª–µ—Ä–∞ –æ —Å—É–º–º–µ —Å—Ç–µ–ø–µ–Ω–µ–π üå∂Ô∏èüå∂Ô∏è

# –í 1769 –≥–æ–¥—É –õ–µ–æ–Ω–∞—Ä–¥ –≠–π–ª–µ—Ä —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–ª –æ–±–æ–±—â–µ–Ω–Ω—É—é –≤–µ—Ä—Å–∏—é –í–µ–ª–∏–∫–æ–π —Ç–µ–æ—Ä–µ–º—ã –§–µ—Ä–º–∞, –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è, —á—Ç–æ –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ n —ç–Ω–Ω—ã—Ö —Å—Ç–µ–ø–µ–Ω–µ–π
# –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å—É–º–º—ã, –∫–æ—Ç–æ—Ä–∞—è —Å–∞–º–∞ —è–≤–ª—è–µ—Ç—Å—è —ç–Ω–Ω–æ–π —Å—Ç–µ–ø–µ–Ω—å—é –¥–ª—è n>2. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –æ–ø—Ä–æ–≤–µ—Ä–∂–µ–Ω–∏—è –≥–∏–ø–æ—Ç–µ–∑—ã –≠–π–ª–µ—Ä–∞ 
# (–ø—Ä–æ–¥–µ—Ä–∂–∞–≤—à–µ–π—Å—è –¥–æ 1967 –≥–æ–¥–∞), –∏ –Ω–∞–π–¥–∏—Ç–µ —á–µ—Ç—ã—Ä–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, —Å—É–º–º–∞ 5-—Ö —Å—Ç–µ–ø–µ–Ω–µ–π –∫–æ—Ç–æ—Ä—ã—Ö —Ä–∞–≤–Ω–∞ 5-–π —Å—Ç–µ–ø–µ–Ω–∏ –¥—Ä—É–≥–æ–≥–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ü–µ–ª–æ–≥–æ —á–∏—Å–ª–∞.

# –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –Ω–∞–π–¥–∏—Ç–µ –ø—è—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª a,b,c,d,e —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö —É—Å–ª–æ–≤–∏—é: a^5+b^5+c^5+d^5=e^5.
 
# –í –æ—Ç–≤–µ—Ç–µ —É–∫–∞–∂–∏—Ç–µ —Å—É–º–º—É a+b+c+d+e.
 
# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª for.
 
# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –°—á–∏—Ç–∞–π—Ç–µ, —á—Ç–æ —á–∏—Å–ª–∞ a,b,c,d,e –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥—è—Ç 150.
 
# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 3. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –º–æ–∂–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –¥–æ–ª—å—à–µ —á–µ–º –æ–±—ã—á–Ω–æ. 
# –í –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Å–ø–æ—Å–æ–±–∞ —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –Ω–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–≥—Ä–∞–º–º—ã –º–æ–∂–µ—Ç —É—Ö–æ–¥–∏—Ç—å –¥–æ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –º–∏–Ω—É—Ç. 
# –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–æ–∫—Ä–∞—Ç–∏—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö —Ü–∏–∫–ª–æ–≤. 

counter = 0
qa = 1
qb = 1
qc = 1
qd = 1
qe = 1
for a in range(1, 151):
    qa = a ** 5
    for b in range(a, 151):
        qb = b ** 5
        for c in range(b, 151):
            qc = c ** 5
            for d in range(c, 151):
                qd = d ** 5
                for e in range(1, 151):
                    qe = e ** 5
                    if qa + qb + qc + qd == qe:
                        counter += 1
                        print("a=", a, "b=", b, "c=", c, "d=", d, "e=", e)
                        print("a+b+c+d+e=", a+b+c+d+e)
                        break
```

```python
# –ó–∞–¥–∞—á–∞_11

# –î–µ–ª–∏—Ç–µ–ª–∏-1 üå∂Ô∏è

# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ a –∏ b (a< b). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∏–∑ –æ—Ç—Ä–µ–∑–∫–∞ [a;b] —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º–æ–π –¥–µ–ª–∏—Ç–µ–ª–µ–π.
 
# –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.
 
# –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –¥–≤–∞ —á–∏—Å–ª–∞ –Ω–∞ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã—Ö –ø—Ä–æ–±–µ–ª–æ–º: —á–∏—Å–ª–æ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º–æ–π –¥–µ–ª–∏—Ç–µ–ª–µ–π –∏ —Å—É–º–º—É –µ–≥–æ –¥–µ–ª–∏—Ç–µ–ª–µ–π.

# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö —á–∏—Å–µ–ª –Ω–µ—Å–∫–æ–ª—å–∫–æ, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∏–∑ –Ω–∏—Ö.

a = int(input())
b = int(input())
maxdel = 0                          # —á–∏—Å–ª–æ —Å –∞–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º–æ–π –¥–µ–ª–∏—Ç–µ–ª–µ–π
maxsum = 0                          # –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Å—É–º–º–∞ –¥–µ–ª–∏—Ç–µ–ª–µ–π
s = 0                               # —Å—É–º–º–∞ –¥–µ–ª–∏—Ç–µ–ª–µ–π

for num in range(a, b+1):
    s = 0                           # —Å–±—Ä–æ—Å –∑–Ω–∞—á–µ–Ω–∏—è —Å—É–º–º—ã –¥–µ–ª–∏—Ç–µ–ª–µ–π –ø–æ—Å–ª–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Ü–∏–∫–ª–∞
    d = 0                           # —Å–±—Ä–æ—Å –¥–µ–ª–∏—Ç–µ–ª—è –ø–æ—Å–ª–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Ü–∏–∫–ª–∞
    for d in range (1, num+1):
        if num % d == 0:
            s += d
    if s > maxsum:
        maxsum = s
        maxdel = num
    elif s == maxsum and num > maxdel:
        maxsum = s
        maxdel = num

print(maxdel, maxsum)
```

```python
# –ó–∞–¥–∞—á–∞_12

# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. 
# –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã–≤–æ–¥—è—â—É—é –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –¥–µ–ª–∏–º–æ—Å—Ç–∏ —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ n –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. 
# –í –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–∞–¥–æ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å –æ—á–µ—Ä–µ–¥–Ω–æ–µ —á–∏—Å–ª–æ –∏ —Å—Ç–æ–ª—å–∫–æ —Å–∏–º–≤–æ–ª–æ–≤ ¬´+¬ª, —Å–∫–æ–ª—å–∫–æ –¥–µ–ª–∏—Ç–µ–ª–µ–π —É —ç—Ç–æ–≥–æ —á–∏—Å–ª–∞.

# –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

# –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ n, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

n = int(input())
counter = 0                                 # —Å—á–µ—Ç—á–∏–∫ –¥–µ–ª–∏—Ç–µ–ª–µ–π

for num in range(1, n+1):                   # –ø–µ—Ä–µ–±–æ—Ä —á–∏—Å–µ–ª
    counter = 0                             # –æ–±–Ω—É–ª–µ–Ω–∏–µ —Å—á–µ—Ç—á–∏–∫–∞ –¥–µ–ª–∏—Ç–µ–ª–µ–π
    for d in range(1, n+2):                 # –ø–µ—Ä–µ–±–æ—Ä –¥–µ–ª–∏—Ç–µ–ª–µ–π
        if num % d == 0:
            counter += 1
    print(num, "+" * counter, sep="", end="")
    print()
```
